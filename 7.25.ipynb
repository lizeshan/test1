```python
def score():
    lscore=[int(x) for x in input('Enter scores:').split()]
    best=max(lscore)
    j=0
    for i in lscore:
        if i>=best-10:
            print('Student ',j,' score is ',i,'and grade is A')
            j+=1
        elif i>=best-20:
            print('Student ',j,' score is ',i,'and grade is B')
            j+=1
        elif i>=best-30:
            print('Student ',j,' score is ',i,'and grade is C')
            j+=1
        elif i>=best-40:
            print('Student ',j,' score is ',i,'and grade is D')
            j+=1
        else:
            print('Student ',j,' score is ',i,'and grade is F')
            j+=1
```

```python
import random
def shuffle(lst):
    ll=len(lst)
    for i in range(10):
        la=random.randint(0,ll-1)
        lb=random.randint(0,ll-1)
        lst[la],lst[lb]=lst[lb],lst[la]
    return lst
```

```python
a=[int(x) for x in input('Enter integers :').split()]
a1=shuffle(a)
for i in a:
    print(i,end=' ')
```

```python
def eldup(lst):
    l=len(lst)
    lst1=[]
    for i in range(l):
        if lst[i] not in lst1:
            lst1.append(lst[i])
    return lst1
```

```python
a=[int(x) for x in input('Enter integers :').split()]
a=eldup(a)
print('The distinct numbers are:')
for i in a:
    print(i,end=' ')
```

```python
def pao(lst):
    ll=len(lst)
    for i in range(ll-1):
        for j in range(ll-i-1):
            if lst[j]>lst[j+1]:
                lst[j],lst[j+1]=lst[j+1],lst[j]
    return lst
```

```python
a=[int(x) for x in input('Enter 10 integers list:').split()]
a=pao(a)
for i in range(len(a)):
    print(a[i],end=' ')
```
