```python
import math
class RegularPolygon:
    def __init__(self,n=3,side=1,x=0,y=0):
        self.n=n
        self.side=side
        self.x=x
        self.y=y
    def getn(self):
        return n
    def getside(self):
        return side
    def getx(self):
        return x
    def gety(self):
        return y
    def setn(self,n):
        self.n=n
    def setside(self,side):
        self.side=side
    def setx(self,x):
        self.x=x
    def sety(self,y):
        self.y=y
    def getPerimeter(self):
        return self.side*self.n
    def getArea(self):
        return (self.n*self.side**2)/(4*math.tan(math.pi/self.n))
```

```python
rp1=RegularPolygon()
print(rp1.getPerimeter())
print(rp1.getArea())
rp2=RegularPolygon(6,4)
print(rp2.getPerimeter())
print(rp2.getArea())
rp3=RegularPolygon(10,4,5,7.8)
print(rp3.getPerimeter())
print(rp3.getArea())
```

```python
class LineraEquation:
    def __init__(self,a,b,c,d,e,f):
        self.__a=a
        self.__b=b
        self.__c=c
        self.__d=d
        self.__e=e
        self.__f=f
    def geta(self):
        return self.__a
    def getb(self):
        return self.__b
    def getc(self):
        return self.__c
    def getd(self):
        return self.__d
    def gete(self):
        return self.__e
    def getf(self):
        return self.__f
    def isSolvable(self):
        if self.__a*self.__d-self.__b*self.__c!=0:
            return True
        else:
            return False
    def getx(self):
        return (self.__d*self.__e-self.__b*self.__f)/(self.__a*self.__d-self.__b*self.__c)
    def gety(self):
        return (self.__a*self.__f-self.__e*self.__c)/(self.__a*self.__d-self.__b*self.__c)
    
```

```python
a,b,c,d,e,f=eval(input('enter the num a,b,c,d,e,f:'))
le=LineraEquation(a,b,c,d,e,f)
if le.isSolvable()==False:
    print('这个方程无解')
else:
    print(le.getx(),le.gety())
```

- 8 进阶
- 定义一个子类，一个父类
    - 父类产生随机数
    - 子类一直判定该随机数是奇数还是偶数
        - 该例子实际上就是一个粗浅的分布式

```python
import random
class father:
    def __init__(self):
        self.Random = None
    def create_num(self):
        self.Random = random.randrange(0,100)
        return self.Random
class Son(father):
    def __init__(self):
        father.__init__(self)
    def result(self):
        while True:
            res_num = self.create_num()
            if res_num % 2 == 0:
                print("%d : 是偶数"%self.Random)
            else:
                 print("%d : 是奇数"%self.Random)
```

```python
Son().result()
```
